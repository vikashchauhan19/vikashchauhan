```python
import numpy as np
import matplotlib.pyplot as plt
```

'plt.show()' should be only written at the end of your graph code .it is not necessary to write it after every graph. You can aslo use 'plt.show()' to display multiple graphs at once. 


A simple graph with  sine wave and cosin wave

```python
t = np.linspace(0, 10, 100)
fig = plt.figure( figsize=(20,5))
plt.plot( t, np.sin(t), 'r--', label= 'sin(t)')
plt.plot( t, np.cos(t), 'b-*', label= 'cos(t)')
plt.legend()
plt.show()
```

a figure object is the outermost container for a Matplotlib plot. The figure object contain multiple Axes objects.So, the figure is the  final graphics that may contain one or more Axes.The Axes represent an individual plot. 

```python
fig, ax = plt.subplots(nrows = 1, ncols= 2, figsize=(20, 6))
ax[0].plot(t, np.sin(t), 'r--', label= 'sine wave of t')
ax[1].plot(t, np.cos(t), 'b-*', label= 'cosine wave of t')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols= 1, figsize=(20, 6))
ax[0].plot(t, np.sin(t), 'r--', label= 'sine wave of t')
ax[1].plot(t, np.cos(t), 'b-*', label= 'cosine wave of t')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
x = np.linspace(0,10, 100)
plt.plot(x, np.sin(x), 'g-', label= 'sine wave of t')
plt.xlabel('x values' , fontsize= 14, fontdict={'color': 'green'})
plt.ylabel('y values' , fontsize= 14, fontdict={'color': 'green'})
plt.title('sine wave' , fontsize= 16, fontdict={'color': 'green', 'weight': 'bold',})
plt.show()
```

```python
# visualization
# stem plot
fig= plt.figure(figsize=(20,6))
x= np.linspace(0, 10, 50)
y= np.sin(x)
plt.stem(x, y, linefmt='r:', markerfmt='r*',basefmt='g')
plt.show()
```

```python
#hexbin plot
fig = plt.figure(figsize=(8,6))
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize= 25, cmap='viridis_r')
plt.colorbar()
plt.show()
```

```python
# bar plot
sales= np.array([100,200,300,400,500,420,300,500,600])
months = np.array(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep'])
plt.bar(x = months, height=sales)
plt.show() 
```

```python
# bar chart with text
plt.bar(x= months,height=sales)
for i in range(len(sales)):
    plt.text(x= i,y=10, s= sales[i], ha='center', fontdict={'color':'black'})
plt.show()    
```

```python
plt.barh(y= months, width=sales)
plt.show()
```

```python
plt.bar(x = months, height=sales, color= 'green', edgecolor= 'black', alpha= 1)
```

```python
colors = ['red','green','blue','yellow','orange','purple','brown','pink','gray']
plt.bar(x=months,height=sales,color= colors,edgecolor= 'black', alpha= .5)
```

```python
plt.bar(x=months,height=sales,color= colors,edgecolor= 'black', alpha= .5, hatch='/') 
```

```python
# box plot
fig= plt.figure(figsize=(20,6))
x = np.random.randint(1,10000, (100,3))
plt.boxplot(x, meanline=True, showmeans=True, showfliers=True, notch=True)
plt.show()
```
