```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('data/indian_cities.csv', index_col=0)
```

```python
df
```

```python
df.columns.to_list()
```

# univariate analysis


```python
df['City'].nunique()# number of unique cities
```

```python
df.set_index('City', inplace=True)
```

```python
df['State'].value_counts()
```

```python
df['State'].value_counts().head()
```

```python
df['State'].value_counts().plot(kind='bar', figsize=(20, 5), color='green', title='State wise city count ')
```

```python
df['State'].value_counts().head().plot(kind='pie',autopct='%.2f%%',figsize=(7,5),title='State wise city count')
```

```python
df['Population'].plot(kind='line',figsize=(5,5),color='red')
```

```python
df.set_index('City',inplace=True)
```

```python
df['Population'].plot(kind='line',figsize=(5,5),color='red')
```

```python
df["Population"].head(100).plot(kind='line',figsize=(20,5),style='ro-')
ticks=np.arange(0,100)
plt.xticks(ticks,df.index[ticks],rotation=90)
plt.show()
```

```python
df['Sexratio'].plot(kind='hist',figsize=(20,5))
```

```python
df["Metro_Population"].head(100).plot(kind='line',figsize=(20,5),style='ro--')
ticks=np.arange(0,100)
plt.xticks(ticks,df.index[ticks],rotation=90)
plt.show()
```

```python
df["Population"].head(100).plot(kind='line',figsize=(20,5),style='ro-',logy=True)
ticks=np.arange(0,100)
# draw a line
plt.axhline(y=10**6,color='b',linestyle='--')
plt.xticks(ticks,df.index[ticks],rotation=90)
plt.show()
```

```python
df[["Population",'Metro_Population']].plot(kind='line',figsize=(20,6),logy=True)
```

```python
df['Pop_Diff']=df['Metro_Population']- df['Population']
df.head(n=5)
```

```python
df[['Population','Pop_Diff']].plot(kind='line',figsize=(20,6))
```
